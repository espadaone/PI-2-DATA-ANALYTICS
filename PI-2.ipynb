```python
import numpy as np
import pandas as pd
from pandas_datareader import data
import yfinance as yf
import yahoofinancials as yahoo_get
from yahoofinancials import YahooFinancials
import yahoo_finance as y_f
import pandas_datareader.data as web
import datetime 


```

```python
# Definir la URL de la página de Wikipedia con la tabla de las compañías SP500
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

# Descargar la tabla de las compañías SP500 como un objeto DataFrame de pandas
table = pd.read_html(url, header=0)[0]
```

```python
table
```

```python
table = table.sort_values(by=['Symbol'], ascending=True) 
```

```python
table=table.reset_index(drop=True)
table
```

```python

new = table['Symbol'].tolist()
new
```

```python
for i in range(len(new)):
    if new[i] == 'BRK.B':
        new[i] = 'BRK-B'
    elif new[i] == 'BF.B':
        new[i] = 'BF-B'
```

```python
table['new'] = new
table
```

```python
last_col = table.pop('new')
table.insert(0, 'new', last_col)
table
```

```python
table=table.drop(['Symbol'], axis=1)
```

```python
table = table.rename(columns={'new': 'Symbol'})
table
```

```python
Filter_by_sec=table.drop(['Headquarters Location','Date added','CIK','Founded'],axis=1)
Filter_by_sec
```

```python
filter_gic=Filter_by_sec.drop(['Security','GICS Sub-Industry'],axis=1)
filter_gic
```

```python
Filter_by_sec.to_csv('areas_sec.csv', sep=',')
```

```python
# Crear un diccionario que mapee los sectores a las compañías correspondientes
sector_dict = {}
for sector in table["GICS Sector"].unique():
    sector_dict[sector] = list(table[table["GICS Sector"]==sector]["Symbol"])
```

```python
sector_dict
```

```python
df_sector = pd.DataFrame.from_dict(sector_dict, orient='index')
```

```python
df_sector
```

```python
df_sector=df_sector.transpose()
df_sector
```

```python
df_sector.to_csv('sector_tick.csv', sep=',')
```
